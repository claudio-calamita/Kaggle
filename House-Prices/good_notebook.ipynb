Let's separate Numerical and Categorical variables for Train and Test data

```python
def separateNumericalCategoricalVariables(df_input):
    df_cat = pd.DataFrame()
    df_num = pd.DataFrame()
    for column in df_input.columns:
        if df_input.dtypes[column] == np.object: # Categorical data
            df_cat[column] = df_input[column]
        else:
            df_num[column] = df_input[column] 
            
    return df_cat, df_num
    
df_cat_train, df_num_train = separateNumericalCategoricalVariables(df)
df_cat_test, df_num_test = separateNumericalCategoricalVariables(df_test)
```

## Preprocess Categorical Variables
We do inner join of train and test categorical variables. This kind of analysis introduce a dependency from test set. The problem is on data collection that is not representative of population. By this way the process could not be reproducible with another test set because is changing the data set.

The most correct process should be: for every categorical variable introduce missing column

```python
df_cat_train_filled = df_cat_train.fillna('Missing')
df_cat_train_filled_dummies = pd.get_dummies(df_cat_train_filled)

df_cat_test_filled = df_cat_test.fillna('Missing')
df_cat_test_filled_dummies = pd.get_dummies(df_cat_test_filled)


df_cat_train_final, df_cat_test_final = df_cat_train_filled_dummies.align(df_cat_test_filled_dummies, join='inner', axis=1)  # inner join
```

### Rebuild Dataframe

```python
X = pd.DataFrame()
X[df_cat_train_final.columns] = df_cat_train_final
X[df_num_train_scaled.columns] = df_num_train_scaled


Z = pd.DataFrame()
Z[df_cat_test_final.columns] = df_cat_test_final
Z[df_num_test_scaled.columns] = df_num_test_scaled
```

### Build Model

```python
### Train Test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
###Regression
reg = LinearRegression().fit(X_train, y_train)
```

```python
## Prediction
y_pred = reg.predict(X_test)
```

```python
print(reg.score(X_train, y_train))
print(mean_squared_log_error(abs(y_pred), y_test))
```

```python
y_text_pred = reg.predict(Z)
```

```python
### Print out
output = pd.DataFrame()

output['Id'] = id_test
output['SalePrice'] = y_text_pred

output.to_csv('MySubmission.csv', index = None, sep = ',' )
```

```python
#### Let's try Lasso that pulls down less significant variables. You should optimize the penalty term
reg_lasso = Lasso(alpha=0.1).fit(X_train, y_train)
```

```python
y_pred_lasso = reg_lasso.predict(X_test)
print(reg_lasso.score(X_train, y_train))
print(mean_squared_log_error(y_pred_lasso, y_test))
y_text_pred_lasso = reg_lasso.predict(Z)

### Print out
output_lasso = pd.DataFrame()

output_lasso['Id'] = id_test
output_lasso['SalePrice'] = y_text_pred_lasso

output_lasso.to_csv('MySubmission_lasso.csv', index = None, sep = ',' )
```

```python
#### Optimize Lasso Regression
alpha = np.arange(0., 100, 1)
scores = []
log_square_mean_error = []

for a in alpha:
    reg_lasso_a = Lasso(alpha=a).fit(X_train, y_train)
    y_pred_lasso_a = reg_lasso_a.predict(X_test)
    
    scores.append(reg_lasso_a.score(X_train, y_train))
    log_square_mean_error.append(mean_squared_log_error(y_pred_lasso_a, y_test))

fig, ax = plt.subplots(1,2, figsize=(8,8))
ax[0].plot(alpha, scores)
ax[1].plot(alpha, log_square_mean_error)
```

```python
reg_lasso = Lasso(alpha=alpha[np.argmin(log_square_mean_error)]).fit(X_train, y_train)

y_pred_lasso = reg_lasso.predict(X_test)
print(reg_lasso.score(X_train, y_train))
print(mean_squared_log_error(y_pred_lasso, y_test))
y_text_pred_lasso = reg_lasso.predict(Z)

### Print out
output_lasso = pd.DataFrame()

output_lasso['Id'] = id_test
output_lasso['SalePrice'] = y_text_pred_lasso

output_lasso.to_csv('MySubmission_lasso_optimized.csv', index = None, sep = ',' )
```
